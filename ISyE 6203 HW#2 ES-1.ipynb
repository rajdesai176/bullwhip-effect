```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
L = 4
alpha = [0.0]
std_OR =[0.0]
std_IR =[0.0]
std_OW =[0.0]
std_IW =[0.0]
std_OD =[0.0]
std_ID =[0.0]
std_OM =[0.0]
std_IM =[0.0]

for a in range(1,100):
    alpha.append(a/100)
    
    #Retailer
    
    DR = [0.0, 10.0]
    Dlist = [9, 12, 8, 11]
    for j in range(1, 502):
        DR.extend(Dlist)
    i = 1
    FR = [0, 10.0]
    OR = [10.0]
    IR = [10.0]
    IPR = [10+10*L]

    for k in range(0,2004):
        FR.append(0.0)
    
    for l in range(0,2005):   
        OR.append(0.0)
        IR.append(0.0)
        IPR.append(0.0)
    
    for i in range(1,2004):
        FR[i+1] = (1-alpha[a]) * FR[i] + alpha[a]*DR[i]
    
    for i in range(1,2004):
        if i-L-1 < 0:
            IR[i] = IR[i-1] + 0 - DR[i]
            IPR[i] = IPR[i-1] + OR[i-1] - DR[i]
            OR[i] = max((L+1)*FR[i+1]-IPR[i], 0.0)
        else:    
            IR[i] = IR[i-1] + OR[i-L-1] - DR[i]
            IPR[i] = IPR[i-1] + OR[i-1] - DR[i]
            OR[i] = max((L+1)*FR[i+1]-IPR[i], 0.0)
    
    std_OR.append(np.std(OR[1001:2000]))
    std_IR.append(np.std(IR[1001:2000]))
    
    #Wholesaler
    
    i = 1
    FW = [0, 10.0]
    OW = [10.0]
    IW = [10.0]
    IPW = [10+10*L]


    for k in range(0,2004):
        FW.append(0.0)
        
    for l in range(0,2005):   
        OW.append(0.0)
        IW.append(0.0)
        IPW.append(0.0)

    for i in range(1,2004):
        FW[i+1] = (1-alpha[a]) * FW[i] + alpha[a]*OR[i-1]

    for i in range(1,2004):
        if i-L-1 < 0:
            IW[i] = IW[i-1] + 0 - OR[i-1]
            IPW[i] = IPW[i-1] + OW[i-1] - OR[i-1]
            OW[i] = max((L+1)*FW[i+1]-IPW[i], 0.0)
        else:    
            IW[i] = IW[i-1] + OW[i-L-1] - OR[i-1]
            IPW[i] = IPW[i-1] + OW[i-1] - OR[i-1]
            OW[i] = max((L+1)*FW[i+1]-IPW[i], 0.0)
            
    std_OW.append(np.std(OW[1001:2000]))
    std_IW.append(np.std(IW[1001:2000]))
    
    #Distributor
    
    i = 1
    FD = [0, 10.0]
    OD = [10.0]
    ID = [10.0]
    IPD = [10+10*L]


    for k in range(0,2004):
        FD.append(0.0)
        
    for l in range(0,2005):   
        OD.append(0.0)
        ID.append(0.0)
        IPD.append(0.0)

    for i in range(1,2004):
        FD[i+1] = (1-alpha[a]) * FD[i] + alpha[a]*OW[i-1]

    for i in range(1,2004):
        if i-L-1 < 0:
            ID[i] = ID[i-1] + 0 - OW[i-1]
            IPD[i] = IPD[i-1] + OD[i-1] - OW[i-1]
            OD[i] = max((L+1)*FD[i+1]-IPD[i], 0.0)
        else:    
            ID[i] = ID[i-1] + OD[i-L-1] - OW[i-1]
            IPD[i] = IPD[i-1] + OD[i-1] - OW[i-1]
            OD[i] = max((L+1)*FD[i+1]-IPD[i], 0.0)
            
    std_OD.append(np.std(OD[1001:2000]))
    std_ID.append(np.std(ID[1001:2000]))
    
    #Manufacturer
    
    i = 1
    FM = [0, 10.0]
    OM = [10.0]
    IM = [10.0]
    IPM = [10+10*L]


    for k in range(0,2004):
        FM.append(0.0)
        
    for l in range(0,2005):   
        OM.append(0.0)
        IM.append(0.0)
        IPM.append(0.0)

    for i in range(1,2004):
        FM[i+1] = (1-alpha[a]) * FM[i] + alpha[a]*OD[i-1]

    for i in range(1,2004):
        if i-L-1 < 0:
            IM[i] = IM[i-1] + 0 - OD[i-1]
            IPM[i] = IPM[i-1] + OM[i-1] - OD[i-1]
            OM[i] = max((L+1)*FM[i+1]-IPM[i], 0.0)
        else:    
            IM[i] = IM[i-1] + OM[i-L-1] - OD[i-1]
            IPM[i] = IPM[i-1] + OM[i-1] - OD[i-1]
            OM[i] = max((L+1)*FM[i+1]-IPM[i], 0.0)

    std_OM.append(np.std(OM[1001:2000]))
    std_IM.append(np.std(IM[1001:2000]))


result_orders = {"Alpha":alpha, "STD_Retailer Orders":std_OR,"STD_Distributor Orders":std_OD,"STD_Wholesaler Orders":std_OW,"STD_Manufacturer Orders":std_OM}
result_inventory = {"Alpha":alpha, "STD_Retailer Inventory":std_IR,"STD_Distributor Inventory":std_ID,"STD_Wholesaler Inventory":std_IW,"STD_Manufacturer Inventory":std_IM}
df_result_orders = pd.DataFrame(result_orders)
df_result_inventory = pd.DataFrame(result_inventory)
```

```python
figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(alpha, std_OR, c ='red', marker = 'o', label = "Retailer")
plt.plot(alpha, std_OW, c ='blue', marker = 'D', label = "Wholesaler")
plt.plot(alpha, std_OD, c ='yellow', marker = '*', label = "Distributor")
plt.plot(alpha, std_OM, c ='green', marker = '^', label = "Manufacturer")
plt.xlim(xmin=0)
plt.ylim(ymin=0)
plt.title("ES1 : Order Values v/s alpha for L=4")
plt.xlabel('Alpha')
plt.ylabel('Order Values')
plt.legend()
```
